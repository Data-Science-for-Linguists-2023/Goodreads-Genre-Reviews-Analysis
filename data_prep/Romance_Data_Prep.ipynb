## Book Info

```python
romance_books = pd.read_json('data/Books/goodreads_books_romance.json', lines=True)
romance_books.head()
```

```python
romance_books.country_code.value_counts()
```

```python
romance_df = pd.merge(romance_reviews, romance_books, on='book_id')
print(romance_df.shape)
print(romance_df.columns)
```

## Genre Info

```python
genres_df = pd.read_json('data/goodreads_book_genres_initial.json', lines=True)
genres_df.head()
```

```python
romance_df = pd.merge(romance_df, genres_df)
romance_df['Genre_Tag'] = 'romance'
romance_df.shape
```

## Author Info

```python
authors_df = pd.read_json('data/goodreads_book_authors.json', lines=True)
authors_df.head()
```

```python
romance_authors = list(romance_df.index.map(lambda x: romance_df.authors.iloc[x][0]))
list(romance_authors[0].values())[0]
```

```python
romance_df['author_id'] = romance_df.index.map(lambda x: int(list(romance_authors[x].values())[0]))
```

```python
authors_df = authors_df[['author_id', 'name']]
```

```python
romance_df = pd.merge(romance_df, authors_df, on ='author_id')
print(romance_df.shape)
```

## Language Filtering

```python
english = (romance_df.language_code == 'eng') | (romance_df.language_code == 'en-GB') | (romance_df.language_code == 'en-US') | (romance_df.language_code == 'en-CA')
romance_df = romance_df[english]
romance_df.shape
```

```python
romance_df.head()
```

## Formatting

```python
romance_df.columns
```

```python
romance_short = romance_df[['user_id', 'rating', 'review_text', 'language_code', 'average_rating', 'num_pages', 'publication_year', 'ratings_count', 'title', 'genres', 'Genre_Tag', 'name']]
romance_short.columns = ['User_ID', 'Rating', 'Text', 'Language', 'Avg_Rating', 'Pages', 'Pub_Year', 'Ratings_Count', 'Title', 'Genres', 'Category', 'Author']
romance_short = romance_short[[ 'Text', 'Rating', 'Title', 'Author', 'Category', 'Genres', 'Language', 'Pages', 'Pub_Year', 'Avg_Rating', 'Ratings_Count', 'User_ID']]
romance_short.shape
```

```python
empty = (romance_short.Text != '') & (romance_short.Text != ' ') & (romance_short.Rating != '') & (romance_short.Rating != ' ')
print(romance_short.shape)
print(romance_short[empty].shape)
romance_short = romance_short[empty]
print(romance_short.shape)
```

## Saving Info

```python
os.makedirs('data/genre_pkls', exist_ok=True)  

f2 = open('data/genre_pkls/romance_short.pkl', 'wb')
pickle.dump(romance_short, f2)
f2.close() 
```
